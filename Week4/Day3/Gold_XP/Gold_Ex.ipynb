Exercise 1 : Export A Dataframe To JSON With Orient (Easy)

```python
import kaggle
import pandas as pd
```

```python
data = {
'UserID': range(1, 11),
'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma', 
         'Frank', 'Grace', 'Hannah', 'Ian', 'Jane'],
'Age': [25, 30, 35, 40, 29, 34, 28, 31, 36, 33],
'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix',
         'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose']
}
```

```python
df = pd.DataFrame(data)
```

```python
json_split = df.to_json(orient='split', indent=4)
json_records = df.to_json(orient='records', indent=4)
json_index = df.to_json(orient='index', indent=4)
json_columns = df.to_json(orient='columns', indent=4)
json_values = df.to_json(orient='values', indent=4)
json_table = df.to_json(orient='table', indent=4)
```

```python
print("JSON Split Orientation:\n", json_split, "\n")
print("JSON Records Orientation:\n", json_records, "\n")
print("JSON Index Orientation:\n", json_index, "\n")
print("JSON Columns Orientation:\n", json_columns, "\n")
print("JSON Values Orientation:\n", json_values, "\n")
print("JSON Table Orientation:\n", json_table, "\n")
```

```python
import pandas as pd

# Sample DataFrame
data = pd.DataFrame({
    'Date': pd.date_range('20210101', periods=20),
    'Product': ['Product A', 'Product B', 'Product C'] * 6 + ['Product A', 'Product B'],
    'Quantity': [2, 3, 5, 7, 1] * 4,
    'Revenue': [210.50, 340.60, 450.30, 120.10, 150.75] * 4
})

# Export to Excel with formatting using xlsxwriter
file_path = 'Test1.xlsx'
with pd.ExcelWriter(file_path, engine='xlsxwriter') as writer:
    data.to_excel(writer, sheet_name='Sheet1', index=False)
    workbook = writer.book
    worksheet = writer.sheets['Sheet1']
    
    # Header format
    header_format = workbook.add_format({'bold': True, 'text_wrap': True,
                                         'valign': 'top', 'fg_color': '#D7E4BC', 'border': 1})
    
    for col_num, value in enumerate(data.columns.values):
        worksheet.write(0, col_num, value, header_format)
    
    # Set number formats
    currency_format = workbook.add_format({'num_format': '$#,##0.00'})
    number_format = workbook.add_format({'num_format': '#,##0'})
    worksheet.set_column('D:D', 18, currency_format)  # For Revenue
    worksheet.set_column('C:C', 18, number_format)    # For Quantity

# Auto-adjust columns' width using openpyxl
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter

wb = load_workbook(file_path)
ws = wb['Sheet1']

def auto_adjust_columns_width(ws):
    for col in ws.columns:
        max_length = 0
        column = col[0].column  # Get the column name
        for cell in col:
            try:
                if len(str(cell.value)) > max_length:
                    max_length = len(cell.value)
            except:
                pass
        adjusted_width = (max_length + 2)
        ws.column_dimensions[get_column_letter(column)].width = adjusted_width

auto_adjust_columns_width(ws)
wb.save(file_path)

print("Excel file has been created and formatted.")
```
