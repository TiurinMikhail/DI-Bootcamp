```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('datascience_salaries.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.isnull().sum() # No missing values!
```

```python
df.groupby('experience_level').describe()
```

```python
df.groupby('experience_level')['salary'].agg(['mean', 'median']).reset_index()
```

```python
salary_mean = df.groupby('experience_level')['salary'].agg(['mean','median']).reset_index()
salary_mean
```

```python
salary_mean.index
```

```python
plt.figure(figsize=(10, 6))
salary_mean.plot(kind='bar', color=['skyblue','orange'])
plt.xlabel('Experience Level', fontsize=12)
plt.ylabel('Average Salary', fontsize=12)
plt.title('Average Salary by Experience Level', fontsize=14)

# Rotating x-axis labels for better readability
plt.xticks(salary_mean['experience_level'].index, rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Plot the bar chart for mean salaries
plt.bar(salary_mean['experience_level'], salary_mean['mean'], color='skyblue', label='Mean Salary')

# Plot the bar chart for median salaries with a different color
plt.bar(salary_mean['experience_level'], salary_mean['median'], color='lightgrey', label='Median Salary')

#plt.xlabel('Experience Level', fontsize=12)
plt.ylabel('Salary', fontsize=12)
plt.title('Average and Median Salaries by Experience Level', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add a legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Plot for mean salaries
plt.subplot(1, 2, 1)
plt.bar(salary_mean['experience_level'], salary_mean['mean'], color='skyblue')
plt.xlabel('Experience Level', fontsize=12)
plt.ylabel('Mean Salary', fontsize=12)
plt.title('Mean Salary by Experience Level', fontsize=14)
plt.xticks(rotation=45)

# Plot for median salaries
plt.subplot(1, 2, 2)
plt.bar(salary_mean['experience_level'], salary_mean['median'], color='lightgreen')
plt.xlabel('Experience Level', fontsize=12)
plt.ylabel('Median Salary', fontsize=12)
plt.title('Median Salary by Experience Level', fontsize=14)
plt.xticks(rotation=45)

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Plot the bar chart for mean salaries
salary_mean.plot(kind='bar', color=['skyblue', 'orange'])

# Set the x-axis ticks to be the experience level names
plt.xticks(range(len(salary_mean.index)), salary_mean['experience_level'], rotation=45)

# Add labels and title
plt.xlabel('Experience Level', fontsize=12)
plt.ylabel('Average Salary', fontsize=12)
plt.title('Average Salary by Experience Level', fontsize=14)

# Show plot
plt.tight_layout()
plt.show()
```
