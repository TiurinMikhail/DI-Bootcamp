Exercise 5: Basic Data Analysis With Google Colab


```python
iris_data.describe()
```

```python
import matplotlib.pyplot as plt

plt.bar(iris_data['species'], iris_data['sepal_width'], label='Sepal Width')
plt.xlabel('Species')
plt.ylabel('Width')
plt.title('Width Comparison by Species')
plt.legend()
plt.show()
```

```python
plt.bar(iris_data['species'], iris_data['petal_width'], label='Petal Width')
plt.xlabel('Species')
plt.ylabel('Width')
plt.title('Width Comparison by Species')
plt.legend()
plt.show()
```

For trend analysis, the columns that could be interesting are 'Year' and 'Avg hrs per day sleeping'. These columns provide information about the average hours of sleep per day over different years. By analyzing the trend in average sleep duration over time, we can observe any long-term patterns or changes in sleeping habits.

For group comparison, the columns that could be interesting are 'Age Group', 'Activity', and 'Sex'. These columns allow us to categorize the data into different groups based on age, type of activity, and gender, respectively. We can then compare the average hours of sleep per day among different age groups, types of activities, or genders to identify any significant differences or trends.
