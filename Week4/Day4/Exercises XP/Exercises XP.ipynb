Exercise 4: Outlier Detection And Handling

```python
Q1F = titanic_data['Fare'].quantile(0.25)
Q3F = titanic_data['Fare'].quantile(0.75)
IQRF = Q3F - Q1F
lower_boundF = Q1F - 1.5 * IQRF
upper_boundF = Q3F + 1.5 * IQRF
print(lower_boundF,upper_boundF)
outliers_data_fare = titanic_data[(titanic_data['Fare'] < lower_boundF) | (titanic_data['Fare'] > upper_boundF)]
titanic_data = titanic_data[(titanic_data['Fare'] >= lower_boundF) & (titanic_data['Fare'] <= upper_boundF)]
```

Exercise 7: Data Transformation For Age Feature

```python
import pandas as pd

# Define the bins for age groups
age_bins = [0, 18, 30, 40, 50, 60, 80]  # You can adjust the age ranges as needed
age_labels = ['0-17', '18-29', '30-39', '40-49', '50-59', '60+']

# Bin the 'Age' column into age groups
titanic_data['AgeGroup'] = pd.cut(titanic_data['Age'], bins=age_bins, labels=age_labels, right=False)

# Apply one-hot encoding to the age groups
age_groups_encoded = pd.get_dummies(titanic_data['AgeGroup'], prefix='AgeGroup')

# Concatenate the one-hot encoded age groups with the original DataFrame
titanic_data_encoded = pd.concat([titanic_data, age_groups_encoded], axis=1)


titanic_data_encoded
```
