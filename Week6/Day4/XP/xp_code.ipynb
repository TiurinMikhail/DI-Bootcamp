Exercise 1 : Linear Regression Analysis

Instructions
Objective: Apply linear regression to a dataset and interpret the results.

Task: Given a dataset of housing prices (house_prices) and their corresponding sizes (house_sizes), use linear regression to predict the price of a house given its size.

```python
import numpy as np
from scipy import stats as stats


house_sizes = [50, 70, 80, 100, 120] 
house_prices = [150.000, 200.000, 210.000, 250.000, 280.000] 
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)
```

```python
print("Slope:", slope)
print("Intercept:", intercept)
```

```python
print("r_value:", r_value)
print("p_value:", p_value)
print("std_err:", std_err)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as ply 


plt.scatter(house_sizes, house_prices, color='lightgreen', label='Data Points')

x_reg = np.linspace(min(house_sizes), max(house_sizes), 100) 
y_reg = slope * x_reg + intercept 
plt.plot(x_reg, y_reg, color='orange', label='Regression Line')

plt.title('Linear Regression')
plt.xlabel('House size')
plt.ylabel('House price')
plt.legend()
plt.show()
```

```python
y_90 = slope*90+intercept
print(f'The price for the house tha is 90 square meters: {y_90}')
```

Exercise 2 : Understanding ANOVA
---


Objective: Test understanding of ANOVA and its application.

Task: Three different fertilizers are applied to three separate groups of plants to test their effectiveness. The growth in centimeters is recorded.

Questions:
--

- Perform an ANOVA test on the given data. What are the F-value and P-value?
- Based on the P-value, do the fertilizers have significantly different effects on plant growth?
- Explain what would happen if the P-value were greater than 0.05.

```python
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]
```

```python
f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print("F-value:", f_value)
print("P-value:", p_value)
```

According to the p value < 0.01 (1%-level) we can say that fertilizers have significantly different effects on plant growth

If the P-value were greater than 0.05:

We would fail to reject the null hypothesis.
This would indicate that there is insufficient evidence to conclude that the fertilizers have different effects on plant growth.
In other words, we would not have enough evidence to say that the fertilizers produce significantly different results in terms of plant growth.

Exercise 4 : Correlation Coefficients
--

Calculate the Pearson and Spearman correlation coefficients between two variables in a dataset.

```python
import pandas as pd
```

```python
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})
```

```python
corr, _  = stats.pearsonr(data['age'],data['income'])
print(f'Pearson Correlation Coefficient: {corr:.2f}')
```

```python
corr, _  = stats.spearmanr(data['age'],data['income'])
print(f'Spearman Rank Correlation: {corr:.2f}')
```

If the Pearson correlation coefficient and the Spearman rank correlation coefficient are both 1.00, it means that there is a perfect positive linear relationship between the variables 'age' and 'income'.

The Pearson correlation coefficient measures the linear relationship between two continuous variables. A value of 1.00 indicates a perfect positive linear relationship.

The Spearman rank correlation coefficient measures the degree of monotonic relationship between two variables, not necessarily linear. A value of 1.00 indicates a perfect positive monotonic relationship.

In other words, as the values of the 'age' variable increase, the values of the 'income' variable also increase, and vice versa.

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(data['age'],data['income'])

plt.scatter(data['age'],data['income'], color='lightgreen', label='Data Points')

x_reg = np.linspace(min(data['age']), max(data['age']), 100) 
y_reg = slope * x_reg + intercept 
plt.plot(x_reg, y_reg, color='orange', label='Regression Line')

plt.title('Linear Regression')
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
plt.show()
```

```python
import plotly.graph_objs as go

scatter_trace = go.Scatter(
    x=data['age'],
    y=data['income'],
    mode='markers',
    marker=dict(color='lightgreen'),
    name='Data Points'
)

# Regression line
x_reg = np.linspace(min(data['age']), max(data['age']), 100)
y_reg = slope * x_reg + intercept
line_trace = go.Scatter(
    x=x_reg,
    y=y_reg,
    mode='lines',
    line=dict(color='orange'),
    name='Regression Line'
)

# Layout
layout = go.Layout(
    title='Linear Regression',
    xaxis=dict(title='Age'),
    yaxis=dict(title='Income'),
    legend=dict(x=0.7, y=0.9)
)

# Plot
fig = go.Figure(data=[scatter_trace, line_trace], layout=layout)
fig.show()
```
