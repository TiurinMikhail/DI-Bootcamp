Exercise 2 : Descriptive Statistics


- Given a sample dataset, calculate the mean, median, variance, and standard deviation using SciPy.
- Sample dataset in code:

```python
import scipy.stats
import numpy as np

data = [12, 15, 13, 12, 18, 20, 22, 21]

mean = scipy.stats.tmean(data)
median = np.median(data)
variance = scipy.stats.tvar(data)
std =scipy.stats.tstd(data)

print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std)
```

Exercise 3 : Understanding Distributions

Generate a normal distribution using SciPy with a mean of 50 and a standard deviation of 10. Plot this distribution.

```python
normal_distr = np.random.normal(loc=50, scale=10, size=300)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(normal_distr, label='Values', color='blue')
plt.xlabel('Price')
plt.ylabel('Quantity / Profit')
plt.title('Demand Curve and Profit Curve')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.hist(normal_distr, bins=30, density=True, alpha=0.6, color='g')
```

Exercise 4 : T-Test Application

Perform a T-test on two sets of randomly generated data. Hint: Use numpy.random.normal() to generate the datasets.

```python
data1 = np.random.normal(loc = 10 ,scale =2 , size= 500)
data2 = np.random.normal(loc = 9 ,scale = 1 , size= 500)
```

```python
t_statistic, p_value = scipy.stats.ttest_ind(data1, data2)
```

```python
if p_value < 0.05:
    print('Reject H0. There is statisticaly signifact diffence between data1 and data2')
else:
    print('There is no statistically signifact diffence between data1 and data2')
```

Exercise 5 : Working With DataFrames

Instructions
Task: Import a dataset using Pandas and calculate descriptive statistics for a specific column using SciPy.|

```python
import pandas as pd
import numpy as np
from scipy import stats
df = pd.read_csv('2019.csv')

column_data = df['GDP per capita']

# Calculate descriptive statistics using SciPy

mean = stats.tmean(column_data)
median = np.median(column_data)
variance = stats.tvar(column_data)
std_deviation = stats.tstd(column_data)

# Print the descriptive statistics
print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_deviation)
```
