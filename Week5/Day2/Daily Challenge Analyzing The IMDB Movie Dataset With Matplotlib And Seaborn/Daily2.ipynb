```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('imdb_top_1000.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Meta_score'] = df['Meta_score'].fillna(df.groupby('Genre')['Meta_score'].transform('median'))
```

```python
df = df.fillna({'Meta_score': df['Meta_score'].median()})
```

```python
df
```

```python
df['Released_Year'].unique()
df['Released_Year'] = df['Released_Year'].str.replace('PG','1995')
```

```python
yearly_ratings = df.groupby('Released_Year')['IMDB_Rating'].mean().reset_index()

# Step 2: This step might be optional if your data is already sorted by 'Year'
yearly_ratings.sort_values('Released_Year', inplace=True)
```

```python
plt.figure(figsize=(20, 10))  
plt.plot(yearly_ratings['Released_Year'], yearly_ratings['IMDB_Rating'], marker='o', linestyle='-', color='orange')
plt.title('IMDB Rating Trends Over the Years')  
plt.xlabel('Released Year')  
plt.ylabel('Average IMDB Rating')  
plt.grid(True)  
plt.show()
```

```python
df['Released_Year'] = df['Released_Year'].astype(int)

bins = [1920, 1950, 1970, 1990, 2000, 2010, 2024]


df['bins_year'] = pd.cut(df['Released_Year'], bins)


yearly_ratings = df.groupby('bins_year')['IMDB_Rating'].mean().reset_index()


plt.figure(figsize=(20, 10))  
plt.plot(yearly_ratings['bins_year'].astype(str), yearly_ratings['IMDB_Rating'], marker='o', linestyle='-', color='orange')
plt.title('IMDB Rating Trends Over the Years')  
plt.xlabel('Released Year')  
plt.ylabel('Average IMDB Rating')  
plt.grid(True)  
plt.show()
```

```python
number_by_genre = df['Genre'].value_counts().reset_index()
number_by_genre_over10 = number_by_genre[number_by_genre['count']>20]
```

```python
genre_counts = df['Genre'].value_counts()

# Select the top N most frequent genres
top_genres = genre_counts.nlargest(10)

# Filter the dataframe to include only the top genres
df_filtered = df[df['Genre'].isin(top_genres.index)]

# Create a box plot
plt.figure(figsize=(12, 8))
sns.boxplot(x='Genre', y='IMDB_Rating', data=df_filtered)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.title('Distribution of IMDB Ratings Across Top Genres')
plt.xlabel('Genre')
plt.ylabel('IMDB Rating')
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```
