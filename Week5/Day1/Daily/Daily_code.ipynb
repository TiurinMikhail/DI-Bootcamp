```python
import matplotlib.pyplot as plt
import random as rd
# Creating a 2x2 grid of subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 8))

regions = rd.choices(df_2019['Country or region'].values,k=6)

gdp_per_capita = df_2019[df_2019['Country or region'].isin(regions)]['GDP per capita']

health_life_exp = df_2019[df_2019['Country or region'].isin(regions)]['Healthy life expectancy']

# Plot GDP per Capita
ax1.bar(regions, gdp_per_capita, color='skyblue')
ax1.set_title('GDP per Capita')
ax1.set_xlabel('Region')
ax1.set_ylabel('GDP per Capita')

# Plot Healthy Life Expectancy
ax2.bar(regions, health_life_exp, color='lightgreen')
ax2.set_title('Healthy Life Expectancy')
ax2.set_xlabel('Region')
ax2.set_ylabel('Healthy Life Expectancy')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()


```

```python
import matplotlib.pyplot as plt
import random as rd

# Creating a 2x1 grid of subplots
fig, ax = plt.subplots(figsize=(12, 6))

# Selecting random regions
regions = rd.choices(df_2019['Country or region'].values, k=6)

# Filtering data for selected regions
data = df_2019[df_2019['Country or region'].isin(regions)]

# Plotting GDP per Capita as a line plot
ax.bar(data['Country or region'], data['GDP per capita'], color='skyblue', label='GDP per Capita')

# Plotting Healthy Life Expectancy as a line plot
ax.plot(data['Country or region'], data['Healthy life expectancy'], marker='o', color='red', label='Healthy Life Expectancy')

# Setting labels and title
ax.set_xlabel('Region')
ax.set_ylabel('Value')
ax.set_title('GDP per Capita vs. Healthy Life Expectancy')
ax.legend()

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
