
Data visualization is important in data analysis because it helps us:

Understand Patterns: Visualizing data makes it easier to spot patterns and trends.

Communicate Findings: Visuals make it simpler to share insights with others.

Identify Anomalies: Visualization helps us find outliers and errors in data.

Make Informed Decisions: Visuals aid decision-making by presenting data in a clear way.

Explore Data: Visualization tools let us interactively explore large datasets.

A line graph shows how one variable changes over time. It's useful for tracking trends and patterns in continuous data, especially over time.

Exercise 2 : Creating A Line Plot For Temperature Variation

```python
import pandas as pd

data = {'Temperature':[72, 74, 76, 80, 82, 78, 75],
'Date':['2024-04-01','2024-04-02','2024-04-03','2024-04-04','2024-04-05','2024-04-06','2024-04-07']} 

df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
import matplotlib.pyplot as plt

# Extract day from the 'Date' column

plt.plot(df['Date'].dt.day_name(), df['Temperature'], label='Temperature in F')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Temperature by Day')
plt.legend()
plt.show()
```

Exercise 3 : Visualizing Monthly Sales With A Bar Chart

```python
data = {'Sales Amount':[5000, 5500, 6200, 7000, 7500],
'Date':['2024-04-01','2024-04-02','2024-05-02','2024-06-02','2024-07-02']} 
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
```

```python
plt.bar(df['Date'].dt.month_name(), df['Sales Amount'],color='orange', edgecolor='black')
plt.title('Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.show()
```

Exercise 4 : Data Visualisation

```python
sales_data = pd.read_csv('sales_data.csv')
```

```python
sales_data.head()
```

```python
total = int(sales_data['quantity'].sum())
print(f'Total quantity of products sold by the company: {total}')
```

Identify the category that had the highest revenue and calculate how much revenue it generated.

```python
sales_data.groupby('category')['revenue'].max().nlargest(1).reset_index()
```

```python
revenue_generated_max = sales_data.groupby('category')['revenue'].sum().nlargest(1).reset_index()
revenue_generated_max
```

Determine the average revenue per sale in the dataset.

```python
averaage_revenue_per_sale = int(sales_data['revenue'].mean())
print(f'Average revenue per sale in the dataset: {averaage_revenue_per_sale}')
```

Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).

```python
sales_data['date'] = pd.to_datetime(sales_data['date'])

quarter = sales_data['date'].dt.quarter
quarter_revenue =  sales_data.groupby(quarter)['revenue'].sum().reset_index()

def format_quarter(quarter):
    return f'Q{quarter}'

plt.bar(quarter_revenue['date'].apply(lambda x: format_quarter(x)),quarter_revenue['revenue'],color='orange', edgecolor='black')
plt.title('Sales Data')
plt.xlabel('Quarter')
plt.ylabel('Sales Amount')
plt.show()
```

Exercise 5 : Data Visualisation Using MatPlotLib

```python
X = [x for x in range(-10,11)]
Y = [x**2 for x in X]
Y2 = [x**3 for x in X]
import matplotlib.pyplot as plt

plt.plot(X, Y,label='y = x^2',color='red')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Simple Line Plot with Labels and Title')
plt.legend(loc = 'best')
plt.show()
```

```python
data={
"products":['A', 'B', 'C', 'D'],
"sales":[15, 30, 45, 20]}

dfs = pd.DataFrame(data)

# Plotting the bar chart
plt.bar(dfs['products'], dfs['sales'], color='skyblue')

# Adding labels and title
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Weekly Sales of Products')

# Displaying the plot
plt.show()
```

```python
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates ']
market_share = [40, 30, 20, 10]

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(market_share, labels=fruits, autopct='%1.1f%%')
plt.title('Favorite fruits of a group of people')
plt.show()
```
